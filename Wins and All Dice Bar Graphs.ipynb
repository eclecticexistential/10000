```python
from bokeh.io import show, output_file
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.plotting import figure
from dice import totals

names = ["Player One", "Player Two", "Gambler"]
wins = [totals[0], totals[1], totals[2]]
colors = ["green", "blue", "red"]

source = ColumnDataSource(data=dict(name=names, data=wins, color=colors))

TOOLS="reset, hover, save"

p = figure(x_range=names, title="10,000 Dice Game Winners Out of 100,000", 
           toolbar_location='right', tools=TOOLS, plot_width=800, plot_height=650, y_range=(0, 80000))

p.vbar(x='name', top='data', width=0.6, source=source, line_color="white", fill_color='color')


p.axis.major_label_text_font_size = "13px"

hover = p.select_one(HoverTool)
hover.tooltips = [
    ("Player", "@name"),
    ("Stats", "@data"),
]

p.title.text_font_size = "24px"
p.xgrid.grid_line_color = None
p.sizing_mode = "stretch_both"

output_file("Dice_Winners.html", title="10,000 Dice Game Winners")

show(p)

dice_num = ["Ones", "Twos", "Threes", "Fours", "Fives", "Sixes"]
all_six = [totals[3][0], totals[3][1], totals[3][2], totals[3][3], totals[3][4], totals[3][5]]
colors = ["red", "orange", "green", "blue", "purple", "teal"]

source = ColumnDataSource(data=dict(name=dice_num, data=all_six, color=colors))

TOOLS="reset, hover, save"

p = figure(x_range=dice_num, title="All Six Dice of One Number out of 100,000 Games", 
           toolbar_location='right', tools=TOOLS, plot_width=800, plot_height=650, y_range=(0, 250))

p.vbar(x='name', top='data', width=0.6, source=source, line_color="white", fill_color='color')


p.axis.major_label_text_font_size = "14px"

hover = p.select_one(HoverTool)
hover.tooltips = [
    ("6 Dice of", "@name"),
    ("Total Occurrence", "@data"),
]

p.title.text_font_size = "24px"
p.xgrid.grid_line_color = None
p.sizing_mode = "stretch_both"

output_file("All_Six_Dice.html", title="All Six Dice of One Number")

show(p)
```
